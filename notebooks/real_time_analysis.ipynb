```python
import pandas as pd
import matplotlib.pyplot as plt
from pymongo import MongoClient
import sys
sys.path.append('../requirements/requirements.txt')


# Connect to the NoSQL database
mongo_uri = "mongodb://localhost:27017"
client = MongoClient(mongo_uri)
db = client['log_database']
collection = db['logs']

# Fetch the latest data from the NoSQL database
latest_data = pd.DataFrame(list(collection.find().sort('_id', -1).limit(100)))

# Display the latest data
print(latest_data.head())
```

```python
# Assuming that 'metric1' and 'metric2' are used for anomaly detection
# Display the distribution of the latest metrics
plt.figure(figsize=(10, 6))
plt.hist(latest_data['metric1'], bins=30, alpha=0.7, color='green', label='Metric 1')
plt.hist(latest_data['metric2'], bins=30, alpha=0.7, color='red', label='Metric 2')
plt.title('Latest Metrics Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```
